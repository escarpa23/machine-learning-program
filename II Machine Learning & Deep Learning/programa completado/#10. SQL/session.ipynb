```python
import pandas
```

```python
pandas.read_sql(sql = 'select * from customers', con=con)
```

```python
pandas.read_sql(
    sql = '''
    select *
    from customers
    inner join employees
    on customers.salesRepEmployeeNumber = employees.employeeNumber
    ''', con = con
)
```

```python
pandas.read_sql(
    sql = '''
    select employees.firstName
    from customers
    inner join employees
    on customers.salesRepEmployeeNumber = employees.employeeNumber
    ''', con = con
)
```

```python
pandas.read_sql(
    sql = '''
    select firstName, count(*)
    from customers
    inner join employees
    on customers.salesRepEmployeeNumber = employees.employeeNumber
    group by firstName
    ''', con = con
)
```

```python
pandas.read_sql(
    sql = '''
    select firstName, count(*), sum(amount)
    from customers
    inner join employees
    on customers.salesRepEmployeeNumber = employees.employeeNumber
    inner join payments
    on customers.customerNumber = payments.customerNumber
    group by firstName
    ''', con = con
)
```
